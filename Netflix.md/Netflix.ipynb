```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/zoephilipiyawa/Desktop/ Projects/SQL Project/netflixData.csv')
```

```python
df.head(10)
df
```

```python
df.isnull().sum() #check for null values
```

```python
df['Imdb Score']= df['Imdb Score'].fillna(0) #fill null values
```

```python
df = df.fillna('No response')
```

```python
df.isnull().sum() 
```

```python
df = df.drop_duplicates() #drop duplicates
df
```

```python
# year with the most releases
top_release = df.groupby('Release Date')['Show Id'].count().sort_values(ascending = False).head(10)
plt.figure(figsize = (10,6))
top_release.plot(kind = 'bar')
plt.xlabel('Release Date')
plt.ylabel('Number of Movies')
plt.title('Year with the Most Releases')

```

```python
# Movies and TV Shows distribution
sns.countplot(x = 'Content Type', data = df)
plt.xlabel('Content Type')
plt.ylabel('Number of Content')
plt.title('Content Type Distribution')
```

```python
#number of tv shows and movies in 2020
df_2020 = df[df['Release Date'] == 2020]
df_2020.groupby('Content Type')['Show Id'].count()
```

```python
# number of tv shows and movies from specific countries
specific_countries = df[df['Production Country'].isin(['United States','United Kingdom','South Korea','Canada'])]
sns.countplot(x ='Production Country', hue = 'Content Type', data = specific_countries)
plt.xlabel('Content Type')
plt.ylabel('Number of Content')
plt.title('Content Type Distribution')
```

```python
df.info()
```

```python
df['Imdb Score'] = df['Imdb Score'].str.split('/').str[0] #split the values and take the first value
df['Imdb Score'] = df['Imdb Score'].astype(float) #convert string to integer
```

```python
df.info()
```

```python
df_filtered_show = df[(df['Imdb Score'] > 8) & (df['Content Type'] == 'TV Show')] 
```

```python
#top 10 tv shows
df_filtered_show[['Title','Genres','Duration','Imdb Score','Content Type']].head(10).sort_values(by = 'Imdb Score',
                                                                                                         ascending = False)
```

```python
df_filtered_movies = df[(df['Imdb Score'] > 8) & (df['Content Type'] == 'Movie')] #top 10 movies
```

```python
df_filtered_movies[['Title','Genres','Duration','Imdb Score','Content Type']].head(10).sort_values(by = 'Imdb Score',
                                                                                                         ascending = False)
```

```python
# popularity between TV SHOWS AND MOVIES
df3 = df.groupby('Content Type')['Imdb Score'].mean()
df3.plot(kind = 'bar')
plt.xlabel('Content Type')
plt.ylabel('Avg Popularity Score')
plt.title('Popularity Rating')
```
